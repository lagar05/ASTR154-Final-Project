```python
# The first step of our project is to use astroquery to query the Vizier data release AllWISE Source Catalog for our sources.
# We will start by obtaining the data for our sample of brown dwarfs, given in N. Skrzypek et al. (2016), 
# see: https://doi.org/10.1051/0004-6361/201527359 for the original article.
# Unfortunately this article only reports some of the WISE photometry, so we will query the AllWISE Source Catalog by coordinate to collect 
# information for these sources. Hence, we must collect the brown dwarf positions (RA & DEC) first from N. Skrzypek et al. (2016):

# Start by defining Vizier object:
v1 = Vizier(columns = ['**'])  # the '**' tells Vizier to list all columns in the catalog UPDATE PROB DONT NEED THIS
v1.ROW_LIMIT=-1  # this tells not to limit the rows returned 

# Next use Vizier.get_catalogs to query Vizier for the catlog we're interested in
bd_catalog = v1.get_catalogs('J/A+A/589/A49')  # J/A+A/589/A49 is the name of the catalog given by N. Skrzypek et al. (2016)

# This catalog contains 3 tables, one with information for the L-dwarfs identitfied in WISE, one for the T-dwarfs, and one for references
# We are interested in the tables containing information for the L & T dwarfs:
ldwarf_table = bd_catalog[0]  # L-dwarf table is 0th table in catalog
tdwarf_table = bd_catalog[1]  # T-dwarf table is 1st table in catalog

filters = ['Jmag', 'e_Jmag', 'Hmag', 'e_Hmag', 'Kmag', 'e_Kmag','W1mag', 'e_W1mag', 'W2mag', 'e_W2mag']  # list to specify columns we want

ldwarf_pos_table = ldwarf_table[filters]  # only grab filter columns from l dwarf table
tdwarf_pos_table = tdwarf_table[filters]  # only grab filter columns from t dwarf table



bd_phot_table = vstack([ldwarf_pos_table, tdwarf_pos_table], metadata_conflicts = 'silent')  # stack these tables together for one table!
bd_phot_table['Label'] = 'BD'  # add column labeling all of these sources as brown dwarf (BD)

# we only want sources with data in all filters! mask out those missing data: 
bd_phot_table = bd_phot_table.to_pandas()  # convert to pandas data frame, easier to work with for this
mask = bd_phot_table[filters].notna().all(axis = 1)  # .notna() grabs only not missing values. .all(axis = 1) checks every column in a row
bd_phot_table = bd_phot_table[mask]  # apply mask
bd_phot_table = Table.from_pandas(bd_phot_table)  # convert back to astropy table for stacking with NBD table later

bd_phot_table  # to check
```

# <center/> MCMC </center>

general description and vibes; how it connects to classification


```python
# these are for testing for now!
app_mag_vals = np.array([14.130, 13.420, 12.810, 12.580, 12.340]) # complete
e_app_mag_vals = np.array([0.010, 0.010, 0.010, 0.020, 0.020]) # complete
wavelength_vals = np.array([1.235, 1.662, 2.159, 3.4, 4.6])  # central wavelengths for J, H, K, W1, W2 filters in microns
F0_vals = np.array([1594, 1024, 666.7, 306.68, 170.66]) # zero point magnitude for each filter in Jy (this is for app mag --> flux conversion)
```

```python
def mag_to_flux(m, e_m, F0):
    F = F0 * 10**(-0.4*m)
    dfdm = -0.4 * np.log(10) * F
    variance_F = (dfdm**2)*(e_m**2)
    e_F = variance_F**(1/2)

    return F, e_F


#note that F_nu = alpha * B_nu where alpha = pi * (R/D)^2 sr. So modeling flux with parameters alpha & temperature
def model_flux(theta, wave):
    T, alpha = theta
    wavelengths = wave * u.micron
    bb = BlackBody(temperature = T * u.K)
    B_lambda = bb(wavelengths)
    F_lambda = (alpha * u.sr) * B_lambda
    F_nu = F_lambda.to(u.Jy, equivalencies = u.spectral_density(wavelengths)).value
    return F_nu
```

```python
# define max & min values for T & alpha
T_min = 500 # complete
T_max = 4000 # complete

# ASK ABOUT ALPHA --> you don't know nothing! use approx radius of these sources & mw distances
R_BD = (0.8 * u.R_jup).to(u.cm)  # average BD radius
dist_min = (1 * u.pc).to(u.cm)  # nearest star ~1pc away
dist_max = (100 * u.pc).to(u.cm)  # ~ scale height MW halo

alpha_min = (np.pi * (R_BD / dist_max)**2).value
alpha_max = (np.pi * (R_BD / dist_min)**2).value

log_alpha_min = np.log(alpha_min)
log_alpha_max = np.log(alpha_max)

def log_prior(theta):
    T, log_alpha = theta

    if T_min <= T <= T_max and log_alpha_min <= log_alpha <= log_alpha_max:
        return 0.
    else:
        return -np.inf
        

    # shape = np.shape(T_arr)
    # T_fin_arr = np.full(shape, 0)
    # alpha_fin_arr = np.full(shape, 0)

    # T_match = (T_min <= T_arr) & (T_arr <= T_max)
    # alpha_match = (alpha_min <= alpha_arr) & (alpha_arr <= alpha_max)

    # T_fin_arr[T_match] = 1
    # alpha_fin_arr[alpha_match] = 1

    # prior = T_fin_arr * alpha_fin_arr
    # ln_prior = np.log(prior)

    # return ln_prior
    

def log_likelihood(theta, app_mag, e_app_mag, F0, wave):  # NEED TO CHECK AXIS STUFF
    T, log_alpha = theta
    alpha = np.exp(log_alpha)
    flux, e_flux = mag_to_flux(app_mag, e_app_mag, F0)
    mu = model_flux((T,alpha), wave)
    
    ln_quant = np.log(2 * np.pi * e_flux**2)
    sq_quant = ((flux - mu) / e_flux)**2
    lnL_arr = -0.5 * (ln_quant + sq_quant)
    lnL = np.sum(lnL_arr)
    return lnL
    
    # temp, a = theta
    # T = temp[...,np.newaxis]
    # alpha = a[..., np.newaxis]

    # flux, e_flux = mag_to_flux(app_mag, e_app_mag, F0)
    # mu = model_flux((T, alpha), wave)

    # ln_quant = np.log(2 * np.pi * e_flux**2)
    # sq_quant = ((flux - mu) / e_flux)**2
    # lnL_arr = -0.5 * (ln_quant + sq_quant)
    # lnL = np.sum(lnL_arr, axis = -1)
    # return lnL


def log_posterior(theta, app_mag, e_app_mag, F0, wave):
    # in the past we were able to just do ln_posterior = logprior + log_likelihood. If logprior returned neg infinity for a value out of the prior
    # bounds (because log(0) --> neg inf), this would get added to whatever the log_liklelihood was and still return neg inf. Here, since the BlackBody
    # function in our log_liklihood cannot take negative temperatures, we must now be extra careful that we are only evaluating log_liklihood within the prior values! So here
    # we construct an if else statement. If the log_prior is returning neg inf (prior = 0), we simply return neg inf and never compute log_likelihood. If 
    # logprior returns a finite value then we are within our prior bounds and can also compute log_liklihood (no negative temperatures!). Then we can add together for
    # log_posterior.
    logprior = log_prior(theta)
    if not np.isfinite(logprior):
        return -np.inf
    else:
        ln_posterior = logprior + log_likelihood(theta, app_mag, e_app_mag, F0, wave)
        return ln_posterior
```

```python
# back-calculate what alpha should be to match observed flux at T=1500K
flux_obs = mag_to_flux(app_mag_vals, e_app_mag_vals, F0_vals)[0]

# model_flux with alpha=1 gives you B_lambda * sr, so:
mu_at_alpha1 = model_flux((1500, 1.0), wavelength_vals)
alpha_needed = np.median(flux_obs / mu_at_alpha1)
print(f"alpha needed: {alpha_needed:.3e}")
print(f"log_alpha_needed: {np.log(alpha_needed):.3f}")
print(f"log_alpha_min: {log_alpha_min:.3f}, log_alpha_max: {log_alpha_max:.3f}")
```

```python
# MCMC

# initialize params
n_params = 2  # number of parameters
n_walkers = 100  # number of walkers
n_steps = 5000  # number of steps

# pick starting positions
T_random = np.random.uniform(low = T_min, high = T_max, size = n_walkers)  # for each walker draw number between T min & max
alpha_random = np.random.uniform(low = log_alpha_min, high = log_alpha_max, size = n_walkers)  # do the same for alpha
initial_guesses = np.array([T_random, alpha_random]).T  # array combining initial guesses for both parameters 
                                                        # note must transpose from (ndim, nwalkers) --> (nwalkers, ndim)
                                                        # so each row is a singler walker's T & alpha values
```

```python
# initialize & run

# initialize sampler
n_dim = np.shape(initial_guesses)[1]  # recall initial_guesses has shape (nwalkers, ndim)

# initialize EnsembleSampler with number of walkers, number of dimensions, and the log_posterior function
sampler = emcee.EnsembleSampler(
                    nwalkers = n_walkers,
                    ndim = n_dim,
                    log_prob_fn = log_posterior,
                    args = [app_mag_vals, e_app_mag_vals, F0_vals, wavelength_vals] 
                    )

# run 
sampler.run_mcmc(initial_guesses, n_steps, progress = True)  # note progress = True adds a progress bar!
```

```python
print(sampler.chain.shape)  # checking this has dimensions (n_walkers, n_steps, n_params) (it does!)

# define function to plot the chains! 
def chain_plots(chain, **kwargs):
    # for T
    plt.figure(figsize = (12,4))  # create figure
    plt.plot(chain[:,:,0].T)  # grab all walkers & their steps with 0th index for T. transpose for number of steps on x axis & walker values on y
    plt.xlabel('Step Number', fontsize = 14)  # add axis labels
    plt.ylabel('Temperature, T', fontsize = 14)
    plt.tick_params(axis = 'both', labelsize = 12) 

    # for alpha
    plt.figure(figsize = (12,4))  # create figure
    plt.plot(chain[:,:,1].T)  # same procedure as for T but grab 1t index for alpha & exponentiate for logalpha --> alpha
    plt.xlabel('Step Number', fontsize = 14)  # add axis labels
    plt.ylabel('Log Alpha, log(α)', fontsize = 14)
    plt.tick_params(axis = 'both', labelsize = 12)

chain_plots(sampler.chain)  # use function!
```

```python
# NEED TO PLAY AROUND WITH SUBTRACTED VALUE

chain_log_probs = sampler.get_log_prob()[-1, :]  # get_log_prob returns (n_steps, n_walkers) so [-1,:] grabs last step for all walkers (should be converged by last step!)
converged_walkers = np.where(chain_log_probs > np.max(chain_log_probs) - 10)[0]  # np.where retunrs tuple so index 0 for indices
chain = sampler.chain[converged_walkers, :,:]
chain_plots(chain)
```

```python
# assessing convergenge 
n_burn = 1000  

chain_plots(chain[:,0:n_burn,:])  # plot the burn in (steps from 0 --> n_burn)
chain_plots(chain[:,n_burn:,:])  # plot the after burn in / convergence (steps from n_burn --> 5000)
```

```python
# T_steps = sampler.flatchain[n_burn * n_walkers:,0]
# alpha_steps = sampler.flatchain[n_burn * n_walkers:,1]

T_steps = chain[:, n_burn:, 0].flatten()
alpha_steps = chain[:, n_burn:,1].flatten()

theta_arr = np.array([T_steps, alpha_steps]).T

labels = ['Temperature, T', 'Log Alpha, log(α)']
corner.corner(theta_arr, labels = labels, show_titles = True)
```
